```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
data = pd.read_csv('input/train.csv')
```

```python
data['DistrictId'] = data['DistrictId'].astype(str)
data.head()
```

```python
data['DistrictId'] = data['DistrictId'].astype(str)
```

```python
data = data.drop(['LifeSquare', 'Healthcare_1', 'DistrictId'], axis = 1)
data = pd.get_dummies(data)
train, valid = train_test_split(data, test_size=0.3, random_state=42)
```

```python
from sklearn.linear_model import LinearRegression as LR
lr = LR()
lr.fit(train.drop(['Price'], axis=1), train['Price'])
```

```python
pred_train = lr.predict(train.drop(['Price'], axis=1))
pred_valid = lr.predict(valid.drop(['Price'], axis=1))
```

```python
from sklearn.metrics import r2_score as r2, mean_squared_error as mse
r2(train['Price'], pred_train), mse(train['Price'], pred_train)
```

```python
r2(valid['Price'], pred_valid), mse(valid['Price'], pred_valid)
```
