## data clean

```python
train.columns
```

```python
train.info()
```

```python
def rooms_clean(df: DF) -> DF:
    _median = train['Rooms'].median()
    df.loc[df['Rooms'] > 6, 'Rooms'] = _median
    df.loc[df['Rooms'] < 1, 'Rooms'] = _median
    return df
```

```python
def square_clean(df: DF) -> DF:
    _median_square = train['Square'].median()
    _median__kitchen = train['KitchenSquare'].median()
    df.loc[df['Square'] < 15, 'Square'] = _median_square
    df.loc[df['Square'] > 250, 'Square'] = _median_square
    df.loc[df['KitchenSquare'] < df['Square']/10, 'KitchenSquare'] = _median__kitchen
    df.loc[df['KitchenSquare'] > df['Square']/2, 'KitchenSquare'] = _median__kitchen
    return df
```

```python
def year_clean(df: DF) -> DF:
    df.loc[df['HouseYear'] > 2020, 'HouseYear'] = 2020
    return df
```

```python
def floor_clean(df: DF) -> DF:
    df.loc[df['Floor'] > 48, 'Floor'] = train['Floor'].median()
    df.loc[df['HouseFloor'] > 48, 'HouseFloor'] = df['Floor']
    df.loc[df['HouseFloor'] == 0, 'HouseFloor'] = df['HouseFloor'].median()
    df.loc[df['HouseFloor'] < df['Floor'], 'HouseFloor'] = df['Floor']
    return df
```

## feature

```python
def add_feat_price(df: DF) -> DF:
    stat1 = train.groupby(['DistrictId', 'Rooms'], as_index=False)[['Price']].mean().rename(columns={'Price': 'mean_price1'})
    stat2 = train.groupby(['Rooms'], as_index=False)[['Price']].mean().rename(columns={'Price': 'mean_price2'})
    mean_price = train['Price'].mean()
    
    df = pandas.merge(df, stat1, on=['DistrictId', 'Rooms'], how='left')
    df = pandas.merge(df, stat2, on=['Rooms'], how='left')
    df['mean_price2'] = df['mean_price2'].fillna(mean_price)
    df['mean_price1'] = df['mean_price1'].fillna(df['mean_price2'])

    return df    
```

```python
def add_ecology_2(df: DF) -> DF:
    df['Ecology_2'] = (df['Ecology_2'] == 'B').astype(int)
    return df
```

```python
def add_ecology_3(df: DF) -> DF:
    df['Ecology_3'] = (df['Ecology_3'] == 'B').astype(int)
    return df
```

```python
def add_shops_2(df: DF) -> DF:
    df['Shops_2'] = (df['Shops_2'] == 'B').astype(int)
    return df
```

```python
def add_feat_floor(df: DF) -> DF:
    return df
```

```python
def data_clean(df: DF) -> DF:
    df = rooms_clean(df)
    df = square_clean(df)
    df = year_clean(df)
    df = add_feat_price(df)
    df = floor_clean(df)
    df = add_ecology_2(df)
    df = add_ecology_3(df)
    df = add_shops_2(df)
    _feats = ['Ecology_1', 
         'Ecology_2',
         'Ecology_3', 
         'Social_1', 
         'Social_2', 
         'Social_3', 
         'Helthcare_2', 
         'Shops_1', 
         'Shops_2']
    for feat in _feats:
        df[f'{feat}^2'] = df[feat]** 2    
    return df
```

## model

```python
from sklearn.ensemble import RandomForestRegressor as RF
```

```python
model = RF(n_estimators=333, max_depth=15, random_state=100, max_features=5, n_jobs=-1)
```

```python
train.columns
```

```python
feats = [
    'DistrictId', 
    'Square', 
    'KitchenSquare',
    'HouseFloor', 
    'HouseYear', 
    'Ecology_1', 
    'Ecology_3', 
    'Helthcare_2', 
    'Shops_1', 
    'Shops_2', 
    'mean_price2',
    'Ecology_1^2',
    'Ecology_3^2', 
    'Social_1^2', 
    'Social_2^2', 
    'Social_3^2',
    'Helthcare_2^2', 
    'Shops_1^2', 
    'Shops_2^2',
        ]
```

```python
model.fit(train.loc[:, feats], train['Price'])
```

```python
DF({'feature': feats, 'importance':model.feature_importances_}).sort_values('importance')
```

```python
from sklearn.metrics import r2_score as r2
```

```python
def evaluate_model(df, model=model, feats=feats, check_r2=True) -> numpy.ndarray:
    prediction = model.predict(df.loc[:, feats])
    
    if check_r2:
        _r2 = r2(df['Price'], prediction)
        print(f'R2: {_r2}')
    return prediction
```

```python
pred_train = evaluate_model(train)
```
